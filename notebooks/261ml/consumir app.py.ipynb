```python
# log_streaming_dstream.py
from pyspark import SparkContext
from pyspark.streaming import StreamingContext

# SparkContext con 2 hilos mínimo para Streaming
sc = SparkContext("local[2]", "LogErrorMonitor")
sc.setLogLevel("ERROR")   # 👈 oculta WARN/INFO de Spark

# Batch interval de 5 segundos
ssc = StreamingContext(sc, 5)

log_dir = "logs"   # carpeta a escuchar

# Cada nuevo archivo que aparezca en logs/ será procesado
lines = ssc.textFileStream(log_dir)

# Filtrar solo líneas con ERROR y sin WARNING
errors = lines.filter(lambda line: ("ERROR" in line) and ("WARNING" not in line))

# Mostrar en consola
errors.pprint()

print(f"Escuchando logs en: {log_dir}")
print("Solo se mostrarán líneas con ERROR (sin WARNING).")

ssc.start()
ssc.awaitTermination()
```
