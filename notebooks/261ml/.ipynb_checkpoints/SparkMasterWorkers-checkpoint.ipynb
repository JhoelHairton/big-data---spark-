```python
#Levantar el server cluster
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("m2m-cluster-test")
    .master("spark://spark-master:7077")
    .config("spark.ui.port", "4040")
    .getOrCreate()
)
spark
```

```python
import random
from time import time

N = 50_000_000
t0 = time()
sc = spark.sparkContext

def inside(_):
    x = random.random()
    y = random.random()
    return 1 if x*x + y*y <= 1.0 else 0

num_partitions = 400
rdd = sc.parallelize(range(N), numSlices=num_partitions)
count_inside = rdd.map(inside).reduce(lambda a, b: a + b)

pi_est = 4.0 * count_inside / N
t1 = time()

print(f"π (cluster) ≈ {pi_est}")
print(f"Tiempo (cluster): {t1 - t0:.2f} segundos")
```
