````python
# Configuración de Spark local
from pyspark.sql import SparkSession

spark = SparkSession.builder \    .appName("clase_spark_local") \    .master("local[*]") \    .config("spark.ui.showConsoleProgress", "true") \    .getOrCreate()

spark
````

````python
# Mini-EDA con Spark
# Ajusta la ruta a tu dataset dentro del contenedor
ruta = "data/tu_archivo.csv"

df = spark.read.csv(ruta, header=True, inferSchema=True)
df.printSchema()
df.show(5)
df.describe().show()
````

````python
# Ejemplo ML con Spark (ajusta columnas según tu dataset)
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import LogisticRegression

# Ajusta estos nombres de columna:
columnas_entrada = ['x1', 'x2', 'x3']
col_label = 'label'

vec = VectorAssembler(
    inputCols=columnas_entrada,
    outputCol='features'
)

df2 = vec.transform(df)

train, test = df2.randomSplit([0.8, 0.2], seed=42)

modelo = LogisticRegression(featuresCol='features', labelCol=col_label)
modelo = modelo.fit(train)

print("Area bajo la curva ROC:", modelo.summary.areaUnderROC)
````
